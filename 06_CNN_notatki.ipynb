## Konwolucyjne sieci neuronowe

* przetwarzanie obrazu z dużą wydajnością, klatek video

**Sieci gęste vs. konwolucyjne**: warstwy gęste (Dense) uczą się cech parametrów globalnie na podst. danych wejściowych, warstwy konwolucyjne uczą się globalnych wzorców. Przykład: głowa kota w lewym dolnym rogu lub prawym górnym - sieci konwolucyjne są niewrażliwe na przesunięcia w przeciwieństwie do sieci gęstych

nawet przy małych zbiorach treningowych są wydajne

uczą się przestrzennej hierarchii wzorców, np. pierwsze warstwy uczą się krawędzi, kolejne wnikają głębiej (np. całe ucho kota)


**Architektura sieci konwolucyjnych**

1. **Input**. Mamy obraz wejściowy (obrazek 128x28px)
2. **Convolution**. Robimy splot (konwolucję) - wydrębniamy cechy tego obrazka i przekazujemy do nast. warstwy jako mapa cech (feature map), map cech każdego obrazka jest dowolna liczba. Czyli tu sieć zaczyna się uczyć lokalnych wzorców. Filtr 4x4 redukuje zdjęcie na 124x124px.
3.Max-Pool. Filtr po obrazie wybiera wartości maksymalne i zwraca 2x mniejszy obrazek (62x62px). 
4. **Dense**. Dane z wszystkich neuronów w Max-Pool przeobrażone w wektor, warstwę ukrytą o zadanej liczbie neuronów (np. 256, czyli dane są w formacie 1x256)
5. **Dense**. Ostatnia warstwa ukryta ma np 128 neuronów (czyli znów mamy wymiar 1x128)
6. **Flaten** - wypłaszcza dane
7. **Output** - 1 węzeł, przewiduje etykietę
